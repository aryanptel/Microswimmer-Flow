Pusher

```python
#pusher

from matplotlib import pyplot as plt

import numpy as np

xi,xf = -10, 10
yi,yf = -10, 10


y, x = np.mgrid[ yi:yf:1000j,xi:xf:1000j]

a = 1
l = 2*a
l_ = (a**2)/l


h_al = ((a/l)**2-(a/l)**4)
g_al = (3/2)*((a/l)-((a/l)**3)/3)
j_al = (1/4)*(a/l)*(1-(a/l)**2)**2

p = np.sqrt(x**2+ y**2)  # -  taken


eta = 1 #define the values
f = 5 #enter the values

cons = f/(8*np.pi*eta)

#2rd equation no constant



#effective terms
S_eff = l*((1-(a/l)**2)*g_al+h_al*(a/l)) #+- outside for the, we took only plus for the pusher

D_eff = -(a**2)*((1-g_al)/3+2*j_al) #D_eff

Q_eff = -(1/2-(g_al/2)*(a/l)**4 + ((a/l)**3)*h_al)*f*l**2 #+ we took for the pusher

#we have taken calculated values of the Qeff term
```

```python
#Q_eff
u1 = Q_eff * ( -15*(x**4)/(p**7) + 12*(x**2)/(p**5) - 1/(p**3))
v1 = Q_eff * (- 15*(x**3)*y/(p**7) + 3*x*y/(p**5) )
```

```python
#D_eff

u2 = (D_eff) * ( 15*(x**2)*(y**2)/(p**7) - 15*(x**2)/(p**5) + 3*(y**2)/(p**5) + 15*(x**4)/(p**7) )
v2 = (D_eff) * ( 15*(x**3)*y/(p**7) + 15*x*(y**3)/(p**7) - 18*x*y/(p**5) )
```

```python
#S_eff
#pm will be there outside the S_eff

u3 = S_eff   *  (-3*(x**3)/(p**5) + x/(p**3) )
v3 = S_eff   *  (-3*(x**2)*y/(p**5) - y/(p**3) )
```

```python
u = u1   +    u2/2    +    u3
v = v1   +    v2/2    +    v3

#figure = plt.figure(figure=(20,20), dpi = 100)
# Create streamline figure
plt.figure(figsize=(5,5))

#plt.arrow(0, 0,-leff/2,0,width = 0.1, color = "red")
#plt.arrow(leff, 0, 0.5*leff,0,width = 0.1, color = "red")
#plt.scatter( 0 , 0 , s = 1000, color="black" )
#plt.scatter( -reff,0 , s = 25, color="red" )

plt.streamplot(x, y, u, v,density = 1)

plt.xlabel('x/a') 
plt.ylabel('y/a') 
plt.title("Pusher for the full flow field")
plt.savefig("Pusher(asytotic flow)")
plt.show()
```
